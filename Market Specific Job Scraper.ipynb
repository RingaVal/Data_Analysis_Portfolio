Libraries

```python
import pandas as pd
from jobspy import scrape_jobs
```

Scraping for "social media manager" or "community manager"

```python
location = "Australia"
search_term = "editor"

jobs = scrape_jobs(
    site_name=["indeed", "linkedin", "zip_recruiter","glassdoor"],
    search_term=search_term,
    location=location,
    is_remote=True,
    results_wanted=500,
    hours_old=336,
    linkedin_fetch_description=True,
    country_indeed='Australia'
)

print(f"Found {len(jobs)} jobs")
print(jobs.head())

filename = f"{location.lower().replace(' ', '_')}_{search_term.lower().replace(' ', '_')}_jobs.xlsx"
jobs.to_excel(filename, index=False)
```
